```python
import numpy as np
import pandas as pd
import rasterio
from GLC19.environmental_raster_glc import PatchExtractor
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
patch_extractor = PatchExtractor("rasters GLC19/", size = 1, verbose = True)
patch_extractor.add_all()
```

```python
def featurize(l):
    return list(map(lambda x: x/16, list(map(sum,list(map(sum,l))))))
```

```python
df = pd.read_csv("testSet.csv",sep=';')
df.head()
```

```python
X = np.asarray(df[["glc19TestOccId","Latitude","Longitude"]])
```

```python
features = []
for _ in X:
    features.append(list(_)+list(patch_extractor[_[1],_[2]]))
```

```python
len(features)
```

```python
labels = df["glc19SpId"]
```

```python
df_X = pd.DataFrame(features)
df_y = pd.DataFrame(labels)
```

```python
dataset = pd.concat([df_X,df_y],axis=1)
```

```python
dataset = dataset.rename({'glc19SpId':99},axis='columns')
```

```python
dataset
```

```python
dataset = dataset.rename(lambda x: x-1, axis = 'columns')
```

```python
dataset
```

```python
dataset = dataset.rename({-1:"Latitude",0:"Longitude",98:"glc19SpId"},axis='columns')
```

```python
dataset.to_csv("size4x4_trusted.csv",index=False)
```
