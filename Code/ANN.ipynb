```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.utils.vis_utils import plot_model
from sklearn.preprocessing import OneHotEncoder
```

```python
cls = Sequential()
```

```python
df = pd.read_csv("size1_trusted.csv")
```

```python
df.head()
```

```python
x = df.iloc[:,:79].values
y = df['glc19SpId'].iloc[:].values
```

```python
enc = OneHotEncoder()
enc.fit(y.reshape(-1,1))
y = enc.fit_transform(y.reshape(-1,1)).toarray()
```

```python
l = enc.get_feature_names()
l = l.tolist()
```

```python
f = []
for i in range(len(l)):
    f.append(l[i].split('_')[1].split('.')[0])
```

```python
cls.add(Dense(256,activation='relu'))
cls.add(Dense(256,activation='relu'))
cls.add(Dropout(rate = 0.02))
cls.add(Dense(256,activation='relu'))
cls.add(Dense(256,activation='relu'))
cls.add(Dropout(rate = 0.02))
cls.add(Dense(256,activation='relu'))
cls.add(Dense(256,activation='relu'))
cls.add(Dense(1348,activation='softmax'))
```

```python
cls.compile(optimizer = 'adam', loss='categorical_crossentropy', metrics = ['accuracy'])
cls.fit(x=x,y=y,epochs = 1)
```

```python
cls.summary()
```

```python
cls.save('ANN_Model_size1.h5')
```

```python
rem = pd.read_csv('praveen.csv',sep = ';')
```

```python
rem = rem[rem['test']==True]
```

```python
IDSel = rem['glc19SpId'].iloc[:].values.tolist()
for i in range(len(IDSel)):
    IDSel[i]=str(IDSel[i])
```

```python
def selecttop30(i):
    global y_pred,IDsel,f
    s = pd.DataFrame(y_pred[i].reshape(-1,1),columns = ['pred'])
    s = pd.concat([s,pd.DataFrame(f,columns = ['classID'])],axis = 1)
    s = s[s['classID'].isin(IDSel)]
    s = s.sort_values(by = ['pred'],axis = 0,ascending = 0)
    s.reset_index(level=0, inplace=True)
    s.reset_index(level=0, inplace=True)
    s = s.drop(['index'],axis = 1)
    s = pd.DataFrame(s.iloc[:30,:].values.tolist(),columns = ['Rank','Probability','ClassID'])
    return s
```

```python
result = pd.DataFrame()
for i in range(len(res)):
    temp = selecttop30(i)
    temp['glc19TestOccId']=t['glc19TestOccId'][i]
    result = pd.concat([result,temp],axis = 0)
```

```python
r = result.copy()
r1 = r.copy()
r = pd.concat([r.drop(['ClassID'],axis  = 1),r1['ClassID']],axis = 1)
r = pd.concat([r.drop(['Rank'],axis  = 1),r1['Rank']],axis = 1)
r = pd.concat([r.drop(['Probability'],axis  = 1),r1['Probability']],axis = 1)
#r = r.drop(['Unnamed: 0'],axis = 1)
```

```python
r['Rank']+=1
r.to_csv('submission1.csv',sep = ';')
```
